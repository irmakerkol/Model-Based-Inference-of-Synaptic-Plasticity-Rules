```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob

for f in sorted(glob.glob('figures/*.png')):
    img = mpimg.imread(f)
    fig, ax = plt.subplots(figsize=(14, 8))
    ax.imshow(img); ax.axis('off')
    ax.set_title(os.path.basename(f), fontsize=14)
    plt.tight_layout(); plt.show()
```

```python
import csv, os
csv_path = 'results/tables1_3.csv'
if os.path.exists(csv_path):
    print(f"{'Plasticity Rule':<35} {'MLP R2_W':>8} {'MLP %Dev':>8} | {'Tay R2_W':>8} {'Tay %Dev':>8}")
    print('-' * 85)
    with open(csv_path) as f:
        for r in sorted(csv.DictReader(f), key=lambda r: float(r['taylor_pde']), reverse=True):
            print(f"{r['rule']:<35} {float(r['mlp_r2w']):>8.2f} {float(r['mlp_pde']):>8.2f} | {float(r['taylor_r2w']):>8.2f} {float(r['taylor_pde']):>8.2f}")
else:
    print("No results yet.")
```
